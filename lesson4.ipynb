```python
#size可以放整數或tuple of ints
#low包含，high不包含，可以先放size再放high
content = np.random.randint(50,high=101,size=(50,5))
```

```python
import pandas as pd
#給index及columns
df1 = pd.DataFrame(data = content,
             index=range(1,51),
             columns=['國文','英文','數學','地理','歷史'])
```

```python
import pandas as pd
#給index及columns
df1 = pd.DataFrame(data = content,
             index=[f"第{i}號" for i in range(1,51)],
             columns=['國文','英文','數學','地理','歷史'])
df1
```

```python
#axis = 0 表直，axis = 1表橫
student_sum = df1.sum(axis = 1)
student_mean = df1.mean(axis = 1)
```

```python
df1['總分'] = student_sum
df1['平均'] = student_mean
df1
```

```python
df2 = df1.sort_values(by = '平均',ascending = False)
df2[:6]
```

```python
df2['排名'] = df1['總分'].rank(method='first',ascending = False)
```
