```python
#size可以放整數或tuple of ints
#low包含，high不包含，可以先放size再放high
content = np.random.randint(50,high=101,size=(50,5))
```

```python
import pandas as pd
#給index及columns
df1 = pd.DataFrame(data = content,
             index=range(1,51),
             columns=['國文','英文','數學','地理','歷史'])
```

```python
#print(df1)
(df1['國文']+df1['英文']+df1['數學']+df1['地理']+df1['歷史'])/5
```
