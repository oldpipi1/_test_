```python
import pandas as pd
info2023 = pd.read_csv('每日各站進出站人數2023.csv')
site2023 = pd.read_csv('車站基本資料集.csv')
site2023
df1 = pd.merge(info2023,site2023,left_on='staCode',right_on='code')
#pd.merge(info2023,site2023,left_on='staCode',right_on='code',how='right')
```

```python
df2 = df1[['trnOpDate','staCode','站名','gateInComingCnt','gateOutGoingCnt','地址']]
df2.columns = ['日期','code','站名','進站','出站','地址']
df2
```
